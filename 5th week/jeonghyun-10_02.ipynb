```
class calculator:
  def __init__(self):
    self.result = 0

  def add(self, a, b):
    self.result = a + b
    return self.result
  
  def minus(self, a, b):
    self.result = a - b
    return self.result

cal = calculator()
print(cal.add(5, 10))
print(cal.minus(25, 85))
```
