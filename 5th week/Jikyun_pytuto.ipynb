## 6. 함수(def)와 클래스(class)

- def 함수이름(입력변수)
- class (여러 함수를 포함) 

```
money = 1000
if money > 5000:
  print("택시를 타고 가자!")
elif money > 1250:
  print("대중교통을 이용하자!")
else:
  print("걸어가라...")

money = 2500
if money > 5000:
  print("택시를 타고 가자!")
elif money > 1250:
  print("대중교통을 이용하자!")
else:
  print("걸어가라...")

money = 10000
if money > 5000:
  print("택시를 타고 가자!")
elif money > 1250:
  print("대중교통을 이용하자!")
else:
  print("걸어가라...")
```

```
def HowToGo(money):
  if money > 5000:
    print("택시를 타고 가자!")
  elif money > 1250:
    print("대중교통을 이용하자!")
  else:
    print("걸어가라...")

HowToGo(1000)
HowToGo(2500)
HowToGo(10000)
```

```
class calculator:
  def __init__(self):
    self.result = 0
  
  def add(self, a, b):
    self.result = a + b
    return self.result
  
  def minus(self, a, b):
    self.result = a - b
    return self.result

cal = calculator()

print(cal.add(4,5))
print(cal.minus(3,9))
```
